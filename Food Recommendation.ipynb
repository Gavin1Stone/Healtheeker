```python
import pandas as pd
```

```python
f1 = "AUSNUT 2011-13 AHS Food Nutrient Database.xls"
f2 = "Release 1 - Food nutrient database.xlsx"

file1 = pd.read_excel(f1)
file2 = pd.read_excel(f2)
```

```python
file2 = pd.ExcelFile(f2)
file2.sheet_names
```

```python
file21 = file2.parse("All solids & liquids per 100g")
file21.head()
```

```python
col = ['Food Name', 'Protein', 'Total dietary fibre', 'Total Fat', 
       'Fructose', 'Glucose', 'Sucrose', 'Maltose', 'Lactose', 'Total sugars'] 
#        'Added sugars', 'Free sugars']
file22 = file21[col].loc[1:,:]
file23 = file22.fillna(0)#.dropna()
file23.head()
```

```python
file23.shape
```

# 1.Fruit

Ackee; Bananas, unripe; Bilberries; Blueberries; Breadfruit; Carambola

Cantaloupe; Cranberry – 1 tbsp; Clementine; Dragon fruit 

Lingonberries; Grapes; Guava, ripe; Honeydew and Galia melons; Kiwifruit;

Lemon including lemon juice; Lime including lime juice; Mandarin; Orange; 

Passion fruit; Paw paw; Papaya; Pineapple

Plantain, peeled; Prickly pear / nopales; Raspberry; Rhubarb

Strawberry; Tamarind; Tangelo


```python
name = 'Raspberry'
f1 = file23[file23['Food Name'].str.contains(name) | file23['Food Name'].str.contains(name.lower())]
if f1.shape[0] != 0:
    fruit += [name] + f1.loc[:,f1.columns[1:]].mean().tolist()

f1
```

```python
name_list = ['Blueberry', 'Cranberry', 'Grape', 'Guava', 'Kiwifruit', 'Lemon', 'Lime', 'Mandarin',
             'Orange', 'Passion', 'Paw', 'Pineapple', 'Prickly', 
             'Rhubarb', 'Strawberry', 'Raspberry', 'Tangelo']

fruit = []
for name in name_list:
    f1 = file23[file23['Food Name'].str.contains(name) & file23['Food Name'].str.contains('raw')]
    if f1.shape[0] != 0:
        fruit += [ [name] + f1.loc[:,f1.columns[1:]].mean().tolist() ]
```

```python
fruit_df = pd.DataFrame(fruit, columns=file23.columns).dropna().round(1)
fruit_df = fruit_df.sort_values(by=['Fructose'], ascending=True).reset_index(drop=True)
fruit_df
```

```python
fruit_df['rate'] = 0
fruit_df.loc[ (fruit_df.Fructose <= 1), 'rate'] = 1
fruit_df.loc[ (fruit_df.Fructose > 1) & (fruit_df.Fructose <= 2), 'rate'] = 0.9
fruit_df.loc[ (fruit_df.Fructose > 2) & (fruit_df.Fructose <= 3), 'rate'] = 0.8
fruit_df.loc[ (fruit_df.Fructose > 3) & (fruit_df.Fructose <= 4), 'rate'] = 0.7
fruit_df.loc[ (fruit_df.Fructose > 4), 'rate'] = 0.6

fruit_df
```

```python
fruit_df.to_csv('fruit rec.csv', index=False) 
```

# 2. Meats, Poultry and Meat Substitutes

Beef; Chicken; Foie gras; Kangaroo; Lamb; Pork; Prosciutto; Quorn, mince; Turkey; Cold cuts / deli meat / cold meats such as ham and turkey breast; Processed meat – check ingredients
 

Fish and Seafood
* Canned tuna
* Fresh fish e.g.
    * Cod; Haddock; Plaice; Salmon; Trout; Tuna
* Seafood (ensuring nothing else is added) e.g.
    * Crab; Lobster; Mussels; Oysters; Prawns; Shrimp



```python
meat = []
name_list = ['Beef', 'Chicken', 'Kangaroo', 'Lamb', 'Pork', 'Turkey', 'Cod', 
        'Salmon', 'Trout', 'Tuna', 'Crab', 'Lobster', 'Oyster', 'Prawn']

for name in name_list:
    f1 = file23[file23['Food Name'].str.contains(name) & file23['Food Name'].str.contains('raw')]
    if f1.shape[0] != 0:
        f1 = f1.sort_values(by=['Total Fat'], ascending=True)
        f2 = f1.head(5)
        meat += [[name] + f1.loc[:,f1.columns[1:]].mean().tolist()]
```

```python
meat_df = pd.DataFrame(meat, columns=file23.columns).dropna().round(1).reset_index(drop=True)
meat_df = meat_df.sort_values(by=['Total Fat'], ascending=True).reset_index(drop=True)
meat_df
```

```python
meat_df['rate'] = 0
meat_df.loc[ (meat_df['Total Fat'] <= 4), 'rate'] = 1
meat_df.loc[ (meat_df['Total Fat'] > 4) & (meat_df['Total Fat'] <= 8), 'rate'] = 0.9
meat_df.loc[ (meat_df['Total Fat'] > 8) & (meat_df['Total Fat'] <= 12), 'rate'] = 0.8
meat_df.loc[ (meat_df['Total Fat'] > 12) & (meat_df['Total Fat'] <= 16), 'rate'] = 0.7
meat_df.loc[ (meat_df['Total Fat'] > 16), 'rate'] = 0.6

meat_df
```

```python
meat_df.to_csv('meat rec.csv', index=False) 
```
